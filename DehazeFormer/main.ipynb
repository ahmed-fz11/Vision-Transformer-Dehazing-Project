```python
import os
import shutil
import random

# Define paths
data_dir = "data/NH-HAZE"
train_dir = "data/NH-HAZE/train"
test_dir = "data/NH-HAZE/test"

# Create train and test directories
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Create GT and hazy subdirectories inside train and test directories
for directory in [train_dir, test_dir]:
    os.makedirs(os.path.join(directory, "GT"), exist_ok=True)
    os.makedirs(os.path.join(directory, "hazy"), exist_ok=True)

# Move images into GT and hazy subdirectories
for i in range(1, 56):
    gt_filename = f"{i:02d}_GT.png"
    hazy_filename = f"{i:02d}_hazy.png"
    
    # Determine whether the image should be in train or test set
    if random.random() < 0.8:  # 80% train, 20% test
        dest_dir = train_dir
    else:
        dest_dir = test_dir

    # Move GT image
    shutil.move(os.path.join(data_dir, gt_filename), os.path.join(dest_dir, "GT", gt_filename))
    # Move hazy image
    shutil.move(os.path.join(data_dir, hazy_filename), os.path.join(dest_dir, "hazy", hazy_filename))
```

```python
#File renaming since NH-HAZE has files name 001_GT.png and 001_hazy.png

import os

def rename_files(directory):
    for filename in os.listdir(directory):
        if "GT" in filename:
            new_filename = filename.replace("GT", "")
            old_path = os.path.join(directory, filename)
            new_path = os.path.join(directory, new_filename)
            os.rename(old_path, new_path)
            print(f"Renamed {filename} to {new_filename}")

def rename_files2(directory):
    for filename in os.listdir(directory):
        if "hazy" in filename:
            new_filename = filename.replace("hazy", "")
            old_path = os.path.join(directory, filename)
            new_path = os.path.join(directory, new_filename)
            os.rename(old_path, new_path)
            print(f"Renamed {filename} to {new_filename}")
# Replace 'directory_path' with the path to your directory containing the files
directory_path = './data/NH-HAZE/test/GT'
directory_path2 = './data/NH-HAZE/test/hazy'
rename_files(directory_path)
rename_files2(directory_path2)


```

```python
#bedde paths
import os
import shutil
import random

# Path to the BeDDe dataset folder
bedde_folder = "BeDDe"

# Output folders
output_train_hazy = "data/Outdoor/train/hazy"
output_train_gt = "data/Outdoor/train/GT"
output_test_hazy = "data/Outdoor/test/hazy"
output_test_gt = "data/Outdoor/test/GT"

# Function to extract fog, gt pairs
def extract_pairs(bedde_folder):
    folders = os.listdir(bedde_folder)
    ind=0
    for folder_name in folders:
        fog_folder = os.path.join(bedde_folder, folder_name, "fog")
        gt_folder = os.path.join(bedde_folder, folder_name, "gt")
        
        fog_files = os.listdir(fog_folder)
        gt_file = os.listdir(gt_folder)[0]  # Assuming there's only one ground truth image
        
        for idx, fog_file in enumerate(fog_files):
            fog_path = os.path.join(fog_folder, fog_file)
            gt_path = os.path.join(gt_folder, gt_file)
            
            
            output_fog_name = f"{ind+1+55}_.png"
            output_gt_name = f"{ind+1+55}_.png"
            ind+=1
            
            yield fog_path, gt_path, output_fog_name, output_gt_name

# Extract fog, gt pairs
pairs = list(extract_pairs(bedde_folder))
print(f"Extracted {len(pairs)} fog, gt pairs")

for i in range(0, 20):
    print(pairs[i])
# Shuffle pairs
random.shuffle(pairs)

# Split into train and test
split_index = int(len(pairs) * 0.8)
train_pairs = pairs[:split_index]
test_pairs = pairs[split_index:]
```

```python

# Move pairs to respective folders
def move_pairs(pairs, output_hazy_folder, output_gt_folder):
    os.makedirs(output_hazy_folder, exist_ok=True)
    os.makedirs(output_gt_folder, exist_ok=True)
    for fog_path, gt_path, output_fog_name, output_gt_name in pairs:
        shutil.copy(fog_path, os.path.join(output_hazy_folder, output_fog_name))
        shutil.copy(gt_path, os.path.join(output_gt_folder, output_gt_name))

move_pairs(train_pairs, output_train_hazy, output_train_gt)
move_pairs(test_pairs, output_test_hazy, output_test_gt)
```
